1.- Crea una clase llamada Raices, donde representaremos los valores de una ecuación de 2º grado. Se tendrá los 3 coeficientes como atributos, llamémosles a, b y c. Hay que insertar estos 3 valores para construir el objeto. (3 Pts.)

Las operaciones que se podrán hacer son las siguientes:

- obtenerRaices(): devuelve las 2 posibles soluciones en forma de cadena.
- obtenerRaiz(): devuelve única raíz, que será cuando solo tenga una solución posible.
- getDiscriminante(): devuelve el valor del discriminante (double), el discriminante tiene la siguiente fórmula, $$b^{2}-4ac$$
- tieneRaices(): devuelve un booleano indicando si tiene dos soluciones, para que esto ocurra, el discriminante debe ser mayor o igual que 0.
- tieneRaiz(): devuelve un booleano indicando si tiene una única solución, para que esto ocurra, el discriminante debe ser igual que 0.
- calcular(): mostrará por consola las posibles soluciones que tiene nuestra ecuación, en caso de no existir solución, mostrarlo también.

Fórmula ecuación 2º grado: 
 $$x = \frac{-b \pm \sqrt{b^{2} - 4ac}}{2a}$$ 

```python


class Raices:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        
    def __obtenerRaices(self):
        x1 = (-self.b - (self.__getDiscriminante())**0.5) / (2 * self.a)
        x2 = (-self.b + (self.__getDiscriminante())**0.5) / (2 * self.a)

        return f'x1={x1:.4}, x2={x2:.4}'
    
    def __obtenerRaiz(self):
        num = -self.b + (self.__getDiscriminante())**0.5
        den = 2 * self.a
        return  f'x = {num / den}'

    def __getDiscriminante(self):
        return self.b**2 - 4*self.a*self.c
    
    def __tieneRaices(self):
        return self.__getDiscriminante() > 0 
    
    def __tieneRaiz(self):
        return self.__getDiscriminante() == 0
    
    def calcular(self):
        sol = 'No tiene raices reales'
        if self.__tieneRaiz():
            sol = self.__obtenerRaiz()
        elif self.__tieneRaices():
            sol = self.__obtenerRaices()
            
        print(sol)
    

def main():
    ecuacion = Raices(1, -1, -2)
    ecuacion.calcular()
    
main()
    
        
```

1.-  Crea una clase llamada Punto con sus dos coordenadas x e y.  (3 Pts.)

- Añade el método inicializador para crear puntos fácilmente. Si no se reciben una coordenada, su valor será cero.
- Sobreescribe el método __str__, para que al imprimir por pantalla un punto aparezca en formato (x,y)
- Añade un método llamado cuadrante que indique a qué cuadrante pertenece el punto, teniendo en cuenta que si x = 0 e y ≠ 0 se sitúa sobre el eje y, si x ≠ 0 e y = 0 se sitúa sobre el eje x y si x = 0 e y = 0 está sobre el origen.
- Añade un método llamado vector, que tome otro punto y calcule el vector resultante entre los dos puntos. El vector se representa como la diferencia entre las coordendas del segundo punto respecto al primero (el segundo menos el primero):AB = (x2-x1, y2-y1) => (5-2, 5-3) => (3,2)
- Añade un método llamado distancia, que tome otro punto y calcule la distancia entre los dos puntos y la muestre por pantalla. La fórmula es la siguiente: d = (x2-x1)^2+(y2-y1)^2



```python
from math import sqrt

class Punto:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def cuadrante(self):
        rpta = ''
        if self.x > 0 and self.y > 0:
            rpta = "{} pertenece al primer cuadrante".format(self)
        elif self.x < 0 and self.y > 0:
            rpta ="{} pertenece al segundo cuadrante".format(self)
        elif self.x < 0 and self.y < 0:
            rpta ="{} pertenece al tercer cuadrante".format(self)
        elif self.x > 0 and self.y < 0:
            rpta ="{} pertenece al cuarto cuadrante".format(self)
        elif self.x != 0 and self.y == 0:
            rpta ="{} se sitúa sobre el eje X".format(self)
        elif self.x == 0 and self.y != 0:
            rpta ="{} se sitúa sobre el eje Y".format(self)
        else:
            rpta ="{} se encuentra sobre el origen".format(self)
        
        return rpta

    def vector(self, otro):
        return "El vector entre {} y {} es ({}, {})".format(self, otro, otro.x - self.x, otro.y - self.y) 

    def distancia(self, otro):
        d = sqrt( (otro.x - self.x)**2 + (otro.y - self.y)**2 )
        return "La distancia entre los puntos {} y {} es {}".format(self, otro, d)
             
    def __str__(self):
        return f'({self.x},{self.y})'

def main():
    A = Punto(2,3)
    B = Punto(5,5)
    C = Punto(-3, -1)
    D = Punto(0,0)

    print(A.cuadrante())
    print(C.cuadrante())
    print(D.cuadrante())

    print(A.vector(B))
    print(B.vector(A))

    print(A.distancia(B))
    print(B.distancia(A))

    print(A.distancia(D))
    print(B.distancia(D))
    print(C.distancia(D))

main()
```
